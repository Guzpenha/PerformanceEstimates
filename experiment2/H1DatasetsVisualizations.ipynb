```r
require("ggplot2")
require("gridExtra")
```

```r
df1 = read.csv("./created_data/hypothesis_data/H1_ml100k.csv")
options(repr.plot.width=6, repr.plot.height=4)
df1 <- within(df1, label <- factor(label, levels=names(sort(table(label))))) # sort
g1 <-ggplot(df1,aes(x=label)) + geom_bar() + coord_flip() + xlab("Model")+ ylab("Number of times it was the best model for a user") + ggtitle("ML100k")
```

```r
df2 = read.csv("./created_data/hypothesis_data/H1_ml1m.csv")
df2 <- within(df2, label <- factor(label, levels=names(sort(table(label))))) # sort
g2<- ggplot(df2,aes(x=label)) + geom_bar() + coord_flip() + xlab("Model")+ ylab("Number of times it was the best model for a user") + ggtitle("ML1M")
```

```r
df3 = read.csv("./created_data/hypothesis_data/H1_ml20m.csv")
df3 <- within(df3, label <- factor(label, levels=names(sort(table(label))))) # sort
g3<- ggplot(df3,aes(x=label)) + geom_bar() + coord_flip() + xlab("Model")+ ylab("Number of times it was the best model for a user") + ggtitle("ML20m")
```

```r
options(repr.plot.width=5, repr.plot.height=5)
grid.arrange(g1,g2,g3)
```
