# SCB Classifier feature importance using all features

```r
library(ggplot2)
library(gridExtra)
df = read.csv("/home/guz/ssd/msc-gustavo-penha/results_analysis/created_data/clf_f_importance.csv")
options(repr.plot.width=8, repr.plot.height=11)
# g1 <- ggplot(df[df$dataset == "yelp",], aes(x= reorder(feature,importance), y=importance, fill=feature_type) ) + 
#   geom_bar(stat = "identity") + coord_flip() +
#   ylab("feature importance") + xlab("Feature") +
#   geom_errorbar(aes(ymin=importance - std, ymax=importance + std), width=.1) + facet_grid(. ~ dataset)

g2 <- ggplot(df[df$dataset == "ml100k",], aes(x= reorder(feature,importance), y=importance, fill=feature_type) ) + 
  geom_bar(stat = "identity") + coord_flip() +
  ylab("feature importance") + xlab("Feature") +
  geom_errorbar(aes(ymin=importance - std, ymax=importance + std), width=.1)+ facet_grid(. ~ dataset)

g2
```
